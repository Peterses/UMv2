```python
# coding=utf-8
# This is a sample Python script.

# Press ⌃R to execute it or replace it with your code.
# Press Double ⇧ to search everywhere for classes, files, tool windows, actions, and settings.

import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

from sklearn.base import clone
from sklearn.metrics import accuracy_score
from sklearn.model_selection import RepeatedStratifiedKFold

from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier

clfs = {
    'GNB': GaussianNB(),
    'SVM': SVC(),
    'kNN': KNeighborsClassifier()
}

datasets = [
    'cleveland',
    'appendicitis',
    'contraceptive',
    'monk-2',
    'bupa',
    'phoneme',
    'newthyroid',
    'hayes-roth',
    'winequality-white',
    'winequality-red',
    'ring',
    'banana',
    'titanic',
    'led7digit',
    'glass',
    'wine',
    'segment',
    'twonorm',
    'texture',
    'spambase'
]
```

```python
# evaluate bagging algorithm for classification
from numpy import mean
from numpy import std
from sklearn.datasets import make_classification
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.ensemble import BaggingClassifier

model = BaggingClassifier(KNeighborsClassifier())

ensembleMethods = {
    'Bagging': BaggingClassifier(),
    'GradientBoosting': GradientBoostingClassifier(),
    'RandomForest': RandomForestClassifier()
}


# evaluate the model
cv = RepeatedStratifiedKFold(
    n_splits=5, 
    n_repeats=2, 
    random_state=1234
)

results = np.zeros((10, len(datasets), len(clfs), len(ensembleMethods)))


for data_id, singleDatset in enumerate(datasets):
    dataset = np.genfromtxt("datasets/%s.dat" % (singleDatset), delimiter=",")
    X = dataset[:, :-1]
    y = dataset[:, -1].astype(int)
    
    for clsf_id, classifier_name in enumerate(clfs):

        for en_id, en_method in enumerate(ensembleMethods):
            n_scores = cross_val_score(
                ensembleMethods[en_method], 
                X, 
                y, 
                scoring='accuracy', 
                cv=cv, n_jobs=-1, 
                error_score='raise'
            )
            
#             print(f'Accuracy: %.3f (%.3f) (dataset: {singleDatset}) (classif: {classifier_name}) (method: {en_method})' % (mean(n_scores), std(n_scores)))
            results[:,data_id,clsf_id,en_id] = n_scores
            print(data_id, clsf_id, en_id)

np.save('results', results)
```

```python
### Wilcoxon
res = np.load('results.npy')

# zakładając ze jest policzone dobrze
mean_res = np.mean(res, axis=0)

mean_res_flat = mean_res.reshape(20,9) # koljeno będą  clf1+ens1, clf1+ens2, clf1+ens3, clf2+ens1 ...
mean_res_flat.shape
```

```python
from scipy.stats import rankdata

ranks = []
for r in mean_res_flat:
    ranks.append(rankdata(r).tolist())
    
ranks = np.array(ranks)

mean_ranks = np.mean(ranks, axis=0)
print("Mean ranks:", mean_ranks)
```

```python
from scipy.stats import ranksums
alpha = 0.05

for i, i_clf in enumerate(clfs):
    for j, j_clf in enumerate(clfs):
        if i == j:
            continue
        stat, p = ranksums(ranks.T[i], ranks.T[j])
        print()
        print("%s vs %s - w_stat: %.3f, p: %.5f" % (i_clf, j_clf, stat, p))
        if p<alpha:
            print('roznica statystyczna')
            if stat>0:
                print('przewaga', i_clf)
            else:
                print('przewaga', j_clf)
```

```python
# T-test
from scipy import stats


res = np.load('results.npy')
print(res.shape)
```

```python
for dataset_id in range(20):
    dataset_res = res[:,dataset_id]
    dataset_res_flat = dataset_res.reshape(10,9).swapaxes(0,1)
    print('----dataset id----', dataset_id)

    for i, i_ds in enumerate(dataset_res_flat):
        for j, j_ds in enumerate(dataset_res_flat):
            if i == j:
                continue
            stat, p = stats.ttest_ind(i_ds,j_ds)
            print("\n%s vs %s - w_stat: %.3f, p: %.5f" % (i+1, j+1, stat, p))
            if p<alpha:
                print('róznica statystyczna')
                if stat>0:
                    print('przewaga', i+1)
                else:
                    print('przewaga', j+1)
```
